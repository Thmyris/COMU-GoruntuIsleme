```python
temp_1=img_1[0,0,:]
temp_1[0]/3+temp_1[1]/3+temp_1[2]/3    #255 in altinda olmasi icin hepsini 3e bolduk, yoksa overflow diyip hata veriyor
#sayinin float olmamasi icin de int yaptik
int(temp_1[0]/3+temp_1[1]/3+temp_1[2]/3)
```

```python
def get_value_from_triple(temp_1):   #3 rengin ortalamasi,
    # temp_1=img_1[0,0,:]
    return int(temp_1[0]/3+temp_1[1]/3+temp_1[2]/3)
get_value_from_triple(img_1[0,0,:])
    
```

```python
m,n,k=img_1.shape
img_1.shape
```

```python
m,n,k=img_1.shape
new_image=np.zeros((m,n), dtype='uint8')
for i in range(m):
    for j in range(n):
        s=get_value_from_triple(img_1[i,k,:])
        new_image[i,j]=s
```

```python
def convert_rgb_to_gray(img_1):
    m,n,k=img_1.shape
    new_image=np.zeros((m,n), dtype='uint8')
    for i in range(m):
        for j in range(n):
            s=get_value_from_triple(img_1[i,k,:])
            new_image[i,j]=s
    return new_image
img_2=convert_rgb_to_gray(img_1)
```

```python
plt.imshow(img_1, cmap='gray')   
plt.show()
plt.imshow(img_2, cmap='gray')   #buraya gray yazdiktan sonra zaten pyplot kendisi gri yapmiyormu? degilmis bak denedik
plt.show()
```

```python
plt.imsave('img_1_grayscaled.png', img_2, cmap='gray')   #cmap ekledik buraya yoksa sari kaydediyordu, hoca cok takilmayin dedi *shrug*, ben takilmadim sizde takilmayin dedi
#3 sayi alip bir sayi dondurceksiniz dedi hoca
# benim gri resim bi garip xd
#paraphrasing: amacimiz kutuphaneyi ogrenmek degil, rotate iicn i ve j yer degistirmistik, burda ise i j nin gosterdigi yerdeki degerleri gri yaptik


# !!!!!!!!!!!!!!!!!!
# The real weighting used for rgb to grayscale conversion is 0.3R+0.6G+0.11B. This is because of the luminosity difference of different wavelengths and this formula produces a "nice" grayscale image instead of a super dark one.
# so averaging is not always correct
# hoca ortalamasini almak yeterli diyo
# !!!!!!!!!!!!!!!!!!
```
